```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/nsethi31/Kaggle-Data-Credit-Card-Fraud-Detection/master/creditcard.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset['Class'].value_counts()
```

* This dataset is totally imbalanced ..

```python
legit = dataset[dataset.Class==0]
Fraud = dataset[dataset.Class==1]
```

```python
legit.shape
```

```python
Fraud.shape
```

##### Create newdataset using undersampling method for legit transaction and then concatenate both fraud and legit .

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

## Now concatenate legit_sample and Fraud

```python
new_dataset = pd.concat([legit_sample,Fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
X = new_dataset.drop(columns='Class')
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(X)
```

```python
X = scaled_data
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y ,test_size=0.1 , stratify=Y,random_state=1 )
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation ..

```python
train_data_pred = model.predict(X_train)
train_data_acc = accuracy_score(train_data_pred,Y_train)
```

```python
print("Accuracy Score is : ",train_data_acc)
```

```python
test_data_pred = model.predict(X_test)
test_data_acc = accuracy_score(test_data_pred,Y_test)
```

```python
print("Accuracy Score is : ",test_data_acc)
```
